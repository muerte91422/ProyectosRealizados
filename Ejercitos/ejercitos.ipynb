```python
import random
class Guerrero():#de donde vienen todos
    def __init__(self,nombre :str,salud :int,fuerza :int,defensa :int,velocidad :int):
        
        if not isinstance(nombre, (str)):
            raise TypeError("El nombre debe ser str")
        
        self.nombre=nombre
        
        if not isinstance(salud, (int)):
            raise TypeError("La salud debe ser int")
        self.salud=salud
        
        if not isinstance(fuerza, (int)):
            raise TypeError("La fuerza debe ser int")
        self.fuerza=fuerza
        
        if not isinstance(defensa, (int)):
            raise TypeError("La defensa debe ser int")
        self.defensa=defensa
        
        if not isinstance(velocidad, (int)):
            raise TypeError("La velocidad debe ser int")
        self.velocidad=velocidad
        
        
    
    def atacar(self,objetivo):
        
        daño=10#numero para poder probar pero que al cambiar de clase se sustituira
        objetivo.salud-=daño
        
    def defender(self,daño):
        
        daño-=self.defensa
    
    def sigue_vivo(self):
        
        if self.salud>0:
            
            return True
        else:
            
            self.velocidad=0#para que asi ya no pueda hacer movimientos ya que esta muerto
            return False
        
    def mostrar_estadisticas(self):
        
        print(f'Las estadisticas de {self.nombre} son:\n \
              Salud     = {self.salud}\n \
              Fuerza    = {self.fuerza}\n \
              Defensa   = {self.defensa}\n \
              Velocidad = {self.velocidad}')   
        
        
#Vienen todas las subclases        

class Caballero(Guerrero):
    clase='caballero'
    critico_arquero=False#para medir cuando esta recibiendo un ataque especial
    def __init__(self,nombre):
        
        self.nombre=nombre
        #los valores son aleatorios segun la clase
        self.salud=random.randint(100,150)
        self.fuerza=random.randint(30,40)
        self.defensa=random.randint(50,70)
        self.velocidad=random.randint(10,20)
        
    def atacar(self,objetivo):
        
        daño=self.fuerza*1.2
        print(f'Daño a realizar {round(daño,2)}')
        return round(daño,2)
        
    def defender(self,daño):
        
        if self.critico_arquero:#caso de que reciba un ataque critico de un arquero
            daño_recibido=daño - ((self.defensa/2) * 0.6)
            
            if daño_recibido<0:#porque repele mas daño del que recibe
                print(f'El daño recibido es 0')
                
                return 0#es el daño recibido
            
            print(f'El daño recibido es: {round(daño_recibido,2)}')
        else:#el resto de los casos    
            daño_recibido=daño - (self.defensa * 0.6)
            
            if daño_recibido<0:#porque repele mas daño del que recibe
                print(f'El daño recibido es 0')
                
                return 0#es el daño recibido
            
            print(f'El daño recibido es: {round(daño_recibido,2)}')
            
        
        return round(daño_recibido,2)#para que nos devuelva cuanto daño recibe realmente
    
class Arquero(Guerrero):
    clase='arquero'
    critico_arquero=False#para medir cuando esta recibiendo un ataque especial
    
    def __init__(self,nombre):
        
        self.nombre=nombre
        #los valores son aleatorios segun la clase
        self.salud=random.randint(70,100)
        self.fuerza=random.randint(25,35)
        self.defensa=random.randint(20,30)
        self.velocidad=random.randint(40,60)
        
    def atacar(self,objetivo):
        
        numero_suerte=random.randint(1,10)#son numeros del uno al diez y por cada numero que se pueda coger son 10 por ciento de que pase
        
        if numero_suerte in [1,2,3]:#30 por ciento porque hay 3
            objetivo.critico_arquero=True
            daño=self.fuerza*1
            print(f'Daño critico a realizar {round(daño,2)}')
            return round(daño,2)
        else:
            daño=self.fuerza*1
            print(f'Daño a realizar {round(daño,2)}')
            return round(daño,2)
    
    def defender(self,daño):
        
        numero_suerte=random.randint(1,10)#son numeros del uno al diez y por cada numero que se pueda coger son 10 por ciento de que pase
        
        if numero_suerte in [1,2]:#20 por ciento porque hay dos
            
            daño_recibido=0
            print(f'El daño recibido es: {round(daño_recibido,2)}')
            return round(daño_recibido,2)
            
        else:
            
            if self.critico_arquero:#caso de que le haga critico un arquero
                daño_recibido=daño-(self.defensa/2)*0.4
                
                if daño_recibido<0:#porque repele mas daño del que recibe
                    print(f'El daño recibido es 0')
                
                    return 0#es el daño recibido
                
                
                print(f'El daño recibido es: {round(daño_recibido,2)}')
                return round(daño_recibido,2)
            
            else:#el resto de los casos
                daño_recibido=daño-self.defensa*0.4
                
                if daño_recibido<0:#porque repele mas daño del que recibe
                    print(f'El daño recibido es 0')
                
                    return 0#es el daño recibido
                
                
                print(f'El daño recibido es: {round(daño_recibido,2)}')
                return round(daño_recibido,2)
            
class Mago(Guerrero):
    clase='mago'
    critico_arquero=False#para medir cuando esta recibiendo un ataque especial
    todos=False#para marcar cuando ataca a todos
    def __init__(self,nombre):
        
        self.nombre=nombre
        #los valores son aleatorios segun la clase
        self.salud=random.randint(50,80)
        self.fuerza=random.randint(40,60)
        self.defensa=random.randint(10,20)
        self.velocidad=random.randint(20,30)
    
    def atacar(self,objetivo):
        
        numero_suerte=random.randint(1,10)#son numeros del uno al diez y por cada numero que se pueda coger son 10 por ciento de que pase
        
        if numero_suerte in [1]:#10 por ciento porque hay uno
            daño=self.fuerza*0.8
            print(f'El daño a realizar a todos {round(daño,2)}')
            todos=True
        else:
            daño=self.fuerza*1.5
            print(f'El daño a realizar {round(daño,2)}')
        return round(daño,2)
    
    def defender(self,daño):
        
        numero_suerte=random.randint(1,10)#son numeros del uno al diez y por cada numero que se pueda coger son 10 por ciento de que pase
        
        if numero_suerte in [1]:#10 por ciento porque hay uno
            daño_recibido=0
            cura=15#vida que recupera a un aliado
            print(f'El daño recibido es: {round(daño_recibido,2)} y cura {cura}')
            return (daño_recibido,cura)
        else:
            if self.critico_arquero:#caso de que haga un critico el arquero
                daño_recibido=daño-(self.defensa/2)*0.4
                
                if daño_recibido<0:#porque repele mas daño del que recibe
                    print(f'El daño recibido es 0')
                
                    return 0#es el daño recibido
                
                
                print(f'El daño recibido es: {round(daño_recibido,2)}')
                return round(daño_recibido,2)
            else:
                daño_recibido=daño-self.defensa*0.4
                
                if daño_recibido<0:#porque repele mas daño del que recibe
                    print(f'El daño recibido es 0')
                
                    return 0#es el daño recibido
                
                
                print(f'El daño recibido es: {round(daño_recibido,2)}')
                return round(daño_recibido,2)
        
class Asesino(Guerrero):
    clase='asesino'
    critico_arquero=False#para medir cuando esta recibiendo un ataque especial
    def __init__(self,nombre):
        
        self.nombre=nombre
        #los valores son aleatorios segun la clase
        self.salud=random.randint(60,90)
        self.fuerza=random.randint(35,45)
        self.defensa=random.randint(15,25)
        self.velocidad=random.randint(50,70)
        
    def atacar(self,objetivo):
        
        numero_suerte=random.randint(1,4)#son numeros del uno al cuatro y por cada numero que se pueda coger son 25 por ciento de que pase
        
        if numero_suerte in [1]:#25 por ciento porque hay uno
            daño=self.fuerza*2
            print(f'El daño critico a realizar {round(daño,2)}')
        else:
            daño=self.fuerza*1
            print(f'El daño a realizar {round(daño,2)}')
        return round(daño,2)
    
    def defender(self,daño):
        
        numero_suerte=random.randint(1,10)#son numeros del uno al diez y por cada numero que se pueda coger son 10 por ciento de que pase
        
        if numero_suerte in [1,2,3]:#30 por ciento porque hay tres
            daño_recibido=0
            print(f'El daño recibido es: {round(daño_recibido,2)}')
            return daño_recibido
        else:
            if self.critico_arquero:#caso de que sufra un critico del arquero
                daño_recibido=daño-(self.defensa/2)*0.15
                
                if daño_recibido<0:#porque repele mas daño del que recibe
                    print(f'El daño recibido es 0')
                
                    return 0#es el daño recibido
                
                print(f'El daño recibido es: {round(daño_recibido,2)}')
                return round(daño_recibido,2)
            else:
                daño_recibido=daño-self.defensa*0.15
                
                if daño_recibido<0:#porque repele mas daño del que recibe
                    print(f'El daño recibido es 0')
                
                    return 0#es el daño recibido
                
                print(f'El daño recibido es: {round(daño_recibido,2)}')
                return round(daño_recibido,2)
                
class Curandero(Guerrero):
    clase='curandero'
    critico_arquero=False#para medir cuando esta recibiendo un ataque especial
    def __init__(self,nombre):
        
        self.nombre=nombre
        #los valores son aleatorios segun la clase
        self.salud=random.randint(80,120)
        self.fuerza=random.randint(10,20)
        self.defensa=random.randint(20,30)
        self.velocidad=random.randint(30,40)
        
    def atacar(self,objetivo):
        daño=self.fuerza*0.8
        print(f'El daño a realizar {daño}')
        return round(daño,2)
    
    def defender(self,daño):
        
        if self.critico_arquero:#caso de que un arquero haga un critico
            daño_recibido=daño-(self.defensa/2)*0.5
            
            if daño_recibido<0:#porque repele mas daño del que recibe
                print(f'El daño recibido es 0')
                
                return 0#es el daño recibido
            
            
            print(f'El daño recibido es: {round(daño_recibido,2)}')
            return round(daño_recibido,2)
        else:
            daño_recibido=daño-self.defensa*0.5
            
            if daño_recibido<0:#porque repele mas daño del que recibe
                print(f'El daño recibido es 0')
                
                return 0#es el daño recibido
            
            
            print(f'El daño recibido es: {round(daño_recibido,2)}')
            return round(daño_recibido,2)
        
    def curar(self,objetivo):
        
        objetivo.salud+=30
        
```
